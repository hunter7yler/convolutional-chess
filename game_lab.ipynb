```python
#pure CNN vs. random moves
from chessmate.simulations import ChessPlayground
from chessmate.engines import Random
from chessmate.utils import display_all_results
from local_engines import CNN
nn = CNN(color=True)
rp = Random()
simulation = ChessPlayground(nn, rp)
simulation.play_multiple_games(100)
display_all_results(simulation.all_results)
```

```python
#Minimax (+ CNN) vs. Minimax (Random)
from chessmate.simulations import ChessPlayground
from chessmate.engines import MiniMax
from chessmate.utils import display_all_results
from local_heuristics import NNP
mmn = MiniMax(color=True, depth=2)
mmn.ordering_heuristic = NNP
mmr = MiniMax(color=False, depth=2)
mmr.move_ordering = False
simulation = ChessPlayground(mmn, mmr)
simulation.play_multiple_games(100)
display_all_results(simulation.all_results)
```
